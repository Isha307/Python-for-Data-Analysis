```python
import pandas as pd
import numpy as np
```

```python
dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv', 'Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,99],
                'Gender':['Male','Female','Female','Male','Male', 'Female','Female']
               }
df=pd.DataFrame(dict1)
df
```

```python
# Find unique values from gender column
df.Gender.unique()
```

```python
# Find no.of unique values from gender column
df.Gender.nunique()
```

```python
# Find count unique values from gender column
df.Gender.value_counts()
```

```python
# find total no. of students having marks between 90 and 100
len(df[(df['Marks']>=90) & (df['Marks']<=100)])
```

```python
sum(df.Marks.between(90,100))
```

```python
# Find average marks
df.Marks.mean()
```

```python
# Apply method
global cnt
cnt = 0
def sum1(num):
  global cnt
  cnt = cnt + num
  return cnt

df['Marks'].apply(sum1)
```

```python
df['Marks']
```

```python
df['Sum'] = df['Marks'].apply(lambda x: sum1(x))
```

```python
# Map Function
df['Gender'] = df['Gender'].map({'Male':0, 'Female':1})
```

```python
df
```

```python
# Drop the column
df.drop('Sum', axis=1, inplace=True)
```
