```python
# 7. Create a Series of length 3 of random number
df = pd.Series(np.random.randn(3))
print(df)

df1 = pd.Series(np.random.randint(low = 1, high = 10, size=3))
print(df1)

df2 = pd.Series(range(3))
print(df2)

df3 = pd.Series(range(1,10,3))
print(df3)
```

```python
# 8. Create a copy in series
data = [1,2,3,4,5]
df = pd.Series(data)
df1 = df.copy() #This is deep copy
df[2] = 7
#print(df,"\n",df1)
df2 = df.copy(deep=False) #this is shallow copy, change in original is reflected in copy also
df[2] = 9
print(df,"\n",df2)

```

```python
# 28. Map in pandas - used for element-wise operations on a Series
df = pd.Series([1,2,3,4])
df1 = df.map(lambda x: x**2)
print(df,"\n",df1)
```

```python
# 28. ApplyMap in pandas - used for element-wise operations on a Dataframe
df = pd.DataFrame([[1, 2], [3, 4], [5, 6]])
df1 = df.applymap(lambda x: x**2)
print(df,"\nNext\n",df1)
```

```python
# 29. Apply: Applying a function along the axis of a DataFrame
df = pd.DataFrame([[1, 2], [3, 4], [5, 6]])
df1 = df.apply(sum) #by default along axis 0
df2 = df.apply(sum, axis=1)
print(df,"\nNext\n",df1,"\nNext\n",df2)
```

```python
# 30. Create a pivot table - summarizes and aggregates data in a tabular format
data = {
    'Date': ['2024-05-01', '2024-05-01', '2024-05-02', '2024-05-02', '2024-05-03'],
    'City': ['New York', 'Los Angeles', 'New York', 'Los Angeles', 'New York'],
    'Temperature': [70, 75, 72, 78, 74],
    'Humidity': [50, 60, 55, 65, 58]
}
df = pd.DataFrame(data)
pivot = df.pivot_table(index='Date', columns='City', values=['Temperature', 'Humidity'])
print(pivot)
```
